```python
import tkinter as tk
import random
import speech_recognition as sr
from gtts import gTTS
import os

beginner_questions = {
    "What is the capital of France?": "Paris",
    "How many continents are there?": "Seven",
    "Which animal is known as the Ship of the Desert?": "Camel"
}

intermediate_questions = {
    "Who painted the Mona Lisa?": "Leonardo da Vinci",
    "Highest dam of India is?": "Tehri Dam",
    "Gir National Park in Gujarat is famous for?": "Lion"
}

advanced_questions = {
    "What is the largest organ in the human body?": "Skin",
    "Who developed the theory of general relativity?": "Albert Einstein",
    "Brain of computer is?": "CPU"
}

def speak_text(text):
    tts = gTTS(text=text, lang='en')
    tts.save("text.mp3")
    os.system("afplay text.mp3") 

def start_quiz(level):
    global current_questions, current_question_index
    current_question_index = 0
    if level == "Beginner":
        current_questions = list(beginner_questions.items())
    elif level == "Intermediate":
        current_questions = list(intermediate_questions.items())
    elif level == "Advanced":
        current_questions = list(advanced_questions.items())
    else:
        return

    random.shuffle(current_questions)
    ask_question()

def ask_question():
    global current_questions, current_question_index
    if current_question_index < len(current_questions):
        question, answer = current_questions[current_question_index]
        label_question.config(text=question)
        speak_text(question)
        current_question_index += 1
    else:
        label_question.config(text="Quiz completed!")

def check_answer():
    global current_question_index
    if current_question_index > 0 and current_question_index <= len(current_questions):
        user_answer = entry_answer.get().strip()
        correct_answer = current_questions[current_question_index - 1][1]
        if user_answer.lower() == correct_answer.lower():
            label_result.config(text="That's correct!", fg="green")
            speak_text("That's correct!")
        else:
            message = f"Sorry, the answer is: {correct_answer}"
            label_result.config(text=message, fg="red")
            speak_text(message)
        entry_answer.delete(0, tk.END)
        ask_question()

def recognize_speech():
    global current_question_index
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        try:
            audio = recognizer.listen(source)
            answer = recognizer.recognize_google(audio).strip()
            entry_answer.delete(0, tk.END)
            entry_answer.insert(0, answer)
        except sr.UnknownValueError:
            label_result.config(text="Sorry, could not understand the audio.", fg="red")
        except sr.RequestError:
            label_result.config(text="Sorry, speech recognition service is unavailable.", fg="red")

root = tk.Tk()
root.title("Quiz Generator App(GK Questions)")

label_instruction = tk.Label(root, text="Select a level to start the quiz:")
label_instruction.pack()

button_beginner = tk.Button(root, text="Beginner", command=lambda: start_quiz("Beginner"))
button_beginner.pack()

button_intermediate = tk.Button(root, text="Intermediate", command=lambda: start_quiz("Intermediate"))
button_intermediate.pack()

button_advanced = tk.Button(root, text="Advanced", command=lambda: start_quiz("Advanced"))
button_advanced.pack()

label_question = tk.Label(root, text="")
label_question.pack()

entry_answer = tk.Entry(root)
entry_answer.pack()

button_check_answer = tk.Button(root, text="Check Answer", command=check_answer)
button_check_answer.pack()

button_speak = tk.Button(root, text="Speak", command=recognize_speech)
button_speak.pack()

label_result = tk.Label(root, text="")
label_result.pack()

current_questions = []
current_question_index = 0

root.mainloop()
```
